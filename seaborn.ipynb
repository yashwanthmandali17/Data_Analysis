# Seaborn - Statistical data visualization Library
## 1. What is Seaborn?

Seaborn is a python library that builds on the top of matplotlib library that makes the statistical data more easy and beautiful

#### Why we need it?

1.For creating violin plots,box plots and pair plots

2.Automatically handles data from pandas datasets.

#### Where we use it?

1.Data analysis and reporting.

2.Exploring relationship between variables.

3.Machine Learning(for visualizing datasets and results).




## 2.importing Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

## 3. Loading & exploring dataset

```python
df = sns.load_dataset("titanic")
df.head()
```

```python
df = sns.load_dataset("titanic")
print(df.tail())
print(df.info())
print(df.describe())
print(df.shape)
df.isnull().sum()
```

## 4. Histograms & Countplots

```python
# Histogram
sns.histplot(df['age'].dropna(), bins=10, kde=True)
plt.title("Histogram of Age")
plt.show()
 
# Countplot
sns.countplot(x='class', data=df)
plt.title("Count of Passengers by Class")
plt.show()
```

### 5. Scatter Plots

```python
sns.scatterplot(x='age', y='fare', data=df, hue='sex')
plt.title("Age vs Fare")
plt.show()
```

### 6. Boxplot & Violin Plot


```python
# Boxplot
sns.boxplot(x='class', y='age', data=df)
plt.title("Boxplot of Age by Class")
plt.show()
 
# Violin plot
sns.violinplot(x='class', y='age', hue='sex', data=df, split=True)
plt.title("Violin Plot of Age by Class and Sex")
plt.show()
```

### 7. Pairplot

```python
iris = sns.load_dataset('iris')
sns.pairplot(iris, hue='species')
plt.show()
```

### 8. Heatmap


```python
# Correlation heatmap
corr = iris.select_dtypes(include=['float64', 'int64']).corr() 
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap of Iris Dataset")
plt.show()
```

### 9. Customization & Themes


```python
sns.set_theme(style="whitegrid")  # options: darkgrid, white, ticks, etc.
sns.histplot(df['age'].dropna(), bins=10, color='orange')
plt.title("Customized Histogram with Seaborn")
plt.show()
```

```python
sns.barplot(x='class', y='fare', data=df, hue='sex', errorbar='sd')
plt.title("Average Fare by Class and Sex")
plt.show()
```

```python
"Show trends or changes in mean values across categories."
sns.pointplot(x='class', y='fare', data=df, hue='sex', markers=['o', 'x'])
plt.title("Point Plot of Fare by Class and Sex")
plt.show()
 
 
sns.kdeplot(df['fare'].dropna(), fill=True, color='teal')
plt.title("KDE Plot of Fare Distribution")
plt.show()
```
