### 2. Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="whitegrid")
```

### 3. Load Dataset


```python
df_flights = sns.load_dataset('flights')
df_flights.head()
```

### 4. Explore Data
 

```python
print("Shape:", df_flights.shape)
print(df_flights.info())
print(df_flights.isnull().sum())
print(df_flights.describe())
```

### Insights:

1. The dataset contains 144 records of monthly passenger numbers from 1949 to 1960.
 
2. There are no missing values.

3. This is a time-series dataset. 

### 5. Analysis with NumPy & Pandas

```python
# Average passengers per year
avg_passengers_by_year = df_flights.groupby('year')['passengers'].mean()
print("Average Passengers by Year:\n", avg_passengers_by_year)
```

### Insights:
 
1. There is a clear and consistent upward trend in the average number of passengers each year.

### 6. Visualization with Matplotlib & Seaborn

```python
# Line plot of average passengers over time
plt.plot(avg_passengers_by_year.index, avg_passengers_by_year.values, marker='o', color='green')
plt.title("Average Passengers by Year")
plt.xlabel("Year")
plt.ylabel("Average Monthly Passengers")
plt.show()
```

### Insights:

1. The line plot visually confirms a strong, almost linear, increase in air travel over the decade.

```python
# Heatmap of passengers by year and month
flights_pivot = df_flights.pivot_table(index='month', columns='year', values='passengers')
sns.heatmap(flights_pivot, cmap='magma', annot=True, fmt=".0f") 
plt.title("Passenger Numbers by Month and Year")
plt.show()
```

### Insights:

1. The heatmap reveals two patterns: a yearly upward trend (getting brighter from left to right) and a seasonal pattern within each year.

2. Passenger numbers consistently peak in the summer months of July and August.
