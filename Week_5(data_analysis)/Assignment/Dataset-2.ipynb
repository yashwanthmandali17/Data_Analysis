### 2. Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="whitegrid")
```

### 3. Load Dataset


```python
df_iris = sns.load_dataset('iris')
df_iris.head()
```

### 4. Explore Data
 

```python
print("Shape:", df_iris.shape)
print(df_iris.info())
print(df_iris.isnull().sum())
print(df_iris.describe())
```

### Insights:
 
1. The dataset contains 150 records of flower measurements across 3 species.
 
2. There are no missing values.
 
3. All features are numerical except for the target variable species.


### 5. Analysis with NumPy & Pandas

```python
# Average petal length
avg_petal_length = np.mean(df_iris['petal_length'])
print("Average Petal Length:", avg_petal_length)

# Average sepal width by species
avg_sepal_width_by_species = df_iris.groupby('species')['sepal_width'].mean()
print("Average Sepal Width by Species:\n", avg_sepal_width_by_species)
```

### Insights:
 
1. The average petal length across all species is approximately 3.76.
 
2. The Setosa species has the highest average sepal width, while Versicolor has the lowest.

### 6. Visualization with Matplotlib & Seaborn

```python
# Boxplot of Petal Length by Species
sns.boxplot(x='species', y='petal_length', data=df_iris)
plt.title("Petal Length Distribution by Species")
plt.show()
```

### Insights:

1. Setosa has a distinctly smaller and less varied petal length than the other two species.

2. Virginica has the largest petal lengths on average, followed closely by Versicolor.

```python
# Pairplot of all features with species hue
sns.pairplot(df_iris, hue='species')
plt.show()
```

### Insights:

1. The pairplot clearly shows that Setosa (blue) is linearly separable from the other two species based on petal length and petal width.

2. Versicolor and Virginica show some overlap but are still largely distinguishable, especially using a combination of features.
